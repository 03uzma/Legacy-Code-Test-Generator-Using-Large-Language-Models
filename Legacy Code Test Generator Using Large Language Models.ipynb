```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch, textwrap
```

```python
model_id = "deepseek-ai/deepseek-coder-6.7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",
    low_cpu_mem_usage=True
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=1500,
    temperature=0.2,
    top_p=0.95,
)
```

```python
output = "Explain what this Java function does:\n\npublic int add(int a, int b){return a+b;}"
out = pipe(output)[0]['generated_text']
print(textwrap.fill(out, width=90))
```

```python
output = """
Generate JUnit test cases for the following Java function.
Make sure to include multiple test scenarios, clear method names, and comments.

Java function:
public int add(int a, int b) {
    return a + b;
}
"""
out = pipe(
    output,
    max_new_tokens=300,
    temperature=0.2,
    do_sample=False
)[0]['generated_text']

print(out)
```

```python
output = """
You are an expert Java software assistant.

For the following Java code:
1. Explain what the function does in simple and detailed language (for documentation).
2. Generate clean, well-commented JUnit test cases to test all major scenarios.

Java function:
public int findMax(int[] nums) {
    int max = nums[0];
    for (int i = 1; i < nums.length; i++) {
        if (nums[i] > max) {
            max = nums[i];
        }
    }
    return max;
}
"""

out = pipe(
    output,
    max_new_tokens=600,
    temperature=0.2,
    do_sample=False
)[0]['generated_text']

print(out)
```

```python
def generate_doc_and_tests(java_code: str):
    prompt = f"""
    You are an expert Java documentation and testing assistant.
    For the following Java code:
    1. Generate detailed **Documentation** describing its functionality, parameters, and behavior.
    2. Generate clear, well-commented **JUnit test cases** covering all major scenarios.

    Java code:
    {java_code}
    """
    return pipe(prompt, max_new_tokens=1500, temperature=0.2, do_sample=False)[0]['generated_text']
```

```python
print(generate_doc_and_tests("public int add(int a, int b){return a+b;}"))
```

```python
import gradio as gr
import os
```

```python
def process_java_file(file):
    # Read file content
    with open(file.name, "r") as f:
        code = f.read()

    # Generate documentation + tests
    result = generate_doc_and_tests(code)
    return result
```

```python
iface = gr.Interface(
    fn=process_java_file,
    inputs=gr.File(file_types=[".java"]),
    outputs=gr.Textbox(
        label="Documentation + JUnit Test Cases",
        lines=30,           # increase number of visible lines
        max_lines=1000,     # optional, allows large scrollable content
        interactive=False,  # read-only
        placeholder="The documentation and JUnit tests will appear here..."
    ),
    title="Java Legacy Code Test & Documentation Generator",
    description="Upload your Java file, and the model will generate both documentation and JUnit test cases."
)

iface.launch(debug=True)
```
